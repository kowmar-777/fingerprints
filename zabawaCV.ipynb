```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('odcisk.png')
kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Orgina')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('wygadzenie')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cv.imwrite("wygladzenie.png", dst)
```

```python
blur = cv.blur(img,(5,5))
plt.subplot(121),plt.imshow(img),plt.title('orgina')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('blur')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cv.imwrite("blur.png", blur)
```

```python
img = cv.imread('blur.png',0)
ret,thresh1 = cv.threshold(img,50,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,70,255,cv.THRESH_BINARY)
ret,thresh3 = cv.threshold(img,90,255,cv.THRESH_BINARY)
ret,thresh4 = cv.threshold(img,100,255,cv.THRESH_BINARY)
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_BINARY)
titles = ['Original Image','50','70','90','100','127']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
